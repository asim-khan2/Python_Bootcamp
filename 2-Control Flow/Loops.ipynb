```python
# while loop

count = 0
while count < 5:
    print(count)
    count += 1
```

```python
count = 0

while count%2==0:
    print(count)
    count += 1
```

```python
# loop control statement

# break
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
# continue

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# pass
for i in range(5):
    if i==3:
        print("the no. is ",i)
        pass
    print(i)
```

```python
# nested loop 
# a loop inside the loop

for i in range(3):
    for j in range(2):
        print(f"i = {i}, j = {j} ")
```

```python
# calculat the sum of first N natural number

num = int(input("Enter the number :"))

sum = 0
for i in range(1,num+1):
    sum += i
print(f"sum of {num} natural number is {sum}")
```

```python
# using while loop
# calculat the sum of first N natural number

num = int(input("Enter the number :"))

sum = 0 
i = 1
while(i <= num):
    sum += i
    i += 1
print(f"sum of {num} natural number is {sum}")
```

```python
# print the prime number 1-100
def isPrime(num):
    for i in range(2,num):
        if num % i == 0:
            return False
    return True

for i in range(2,101):
    bool = isPrime(i)

    if bool == True:
        print(i)
```
