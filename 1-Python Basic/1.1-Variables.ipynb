```python
## Dynamic Typing
## Python allows the type of a vraible to change as the program executes

var = 2 # int
print(var,type(var))

var = "asim" # string
print(var,type(var))

var = 3.12 # float
print(var,type(var))

var = True # boolean
print(var,type(var))
```
