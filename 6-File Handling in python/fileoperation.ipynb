## File Operation- Read And Write Files
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
with open('Example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## read the file line by line
with open('Example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip method remove the line space (character)
```

```python
## Writing the file (Overwriting)

with open('Example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('this is a new line\n')
    file.write('Hello My name is Asim')
```

```python
with open('Example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
## Writing the file (without the overwriting) 'a' = append mode
with open('Example.txt','a') as file:
    file.write("i am a passoniat about the machine learning\n")
    file.write('i have a ml model\n')
```

```python
## print the file
with open('Example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
## writing the list of line to a file

line = ['first liine\n','second line\n','third line\n']

with open('Example.txt','a') as file:
    file.writelines(line)
```
