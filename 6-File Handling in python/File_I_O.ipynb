```python
## problem with 'w'
## introducing the append mode =  'a'

f = open('sample1.txt','a')
f.write('\ni am fine\n')
f.close()
```

```python
## write line
l = ['hello\n','how are you\n','i am fine\n']

f = open('sample1.txt','a')
f.writelines(l)
f.close()
```

```python
## reading from file
f = open('sample.txt','r')
content = f.read()
print(content)
f.close()
```

```python
## reading from file
f = open('sample1.txt','r')
content = f.read(10)
print(content)
f.close()
```

```python
## readline() -----> method read the line only one
f = open('sample1.txt','r')
print(f.readline().strip())  ## strip method remove the all {'\n','\t',' '} form the content
print(f.readline())
f.close()
```

```python
## read the file line by line using the readlines() method read the all line form the file
f = open('sample1.txt','r')
content = f.readlines()
print(content)
f.close()
```

```python
## read entire data using readline
f = open('sample1.txt','r')

while f.readline() != '':
    print(f.readline(),end='')
f.close()
```

## Using Context Manager (with)
- It's a good idea to close a file after usage as it will free up the resources.
- If we don't close it, the garbage collector would close it.
- The with keyword closes the file as soon as the usage is over.

```python
## with
with open('sample1.txt','w') as f:
  f.write('Asim Khan')
```

```python
## read the data usign the with
with open('sample.txt','r') as f:
  content = f.read()
  print(content)
```

```python
## add some data in the file
l = ['hello\n','how are you\n','i am fine\n']

f = open('sample1.txt','a')
f.writelines(l)
f.close()
```

```python
## moving within the file 10 char then 10 char
with open('sample1.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
```

```python
## Benifit?---> to load big file in a memory
big_L = ['hello world' for i in range(1000)]

with open('big.txt','w') as f:
  f.writelines(big_L)
```
