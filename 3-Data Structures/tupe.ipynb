A tuple is an ordered, immutable collection of elements, written inside parentheses () and separated by commas

```python
## creating tuple

empty_tuple  = ()

print(empty_tuple)
print(type(empty_tuple))
```

```python
## list using class list
lis = list()
print(lis)          # []
print(type(lis))    # <class 'list'>

tup = tuple()
print(tup)          # ()
print(type(tup))    # <class 'tuple'>
```

```python
number = tuple([1,2,5,6,7,7]) # list convert into tuple (type Casting)
print(number)
```

```python
print(type(number))

# tuple convert into list
list((1,2,4,556,))
```

```python
mixed_tuple = (3,4,5,"Asim",3.14)
print(mixed_tuple)
```
