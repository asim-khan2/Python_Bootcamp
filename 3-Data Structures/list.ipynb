# Defination

A list in Python is an ordered, mutable collection of items that allows storing multiple values in a single variable.
It is written using square brackets [ ], and the items inside a list can be of different data types such as integers, strings, floats, or even other lists.

A list is mutable, which means its elements can be changed, updated, added, or removed after the list is created.

```python
# collection of different data type write in [] bracket
list = ['Asim','Najim','Salim',19, 2.4, 'A'] 
print(list)
```

```python
list[0] = 'Kasim' # index start at 0 
```

```python
print(list)
```

```python
len(list)
```

```python
# list from the string 
a = list('Asim')
print(type(a))
```

```python
# list from the set()
set = list(("Asim"))
```

```python
print(type(set))
```

```python
# list from a set{}

set = list({10,50,23,6,'A'}) # unorder 
print(set,type(set))
```

```python
# list from the dictionary {'a': 1, 'b' : 3}
dict = {'a' : 1,'b' :3}
print(dict, type(dict))

key = list(dict.keys())
print("Key",key)
items_list = list(dict.items())
value = list(dict.values())
print("values",value)
```
