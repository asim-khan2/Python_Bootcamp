## Encapsulation
Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a means of preventing accidental interference and misuse of the data.

```python
#### #Encapsulation with Getter and Setter Method
## Public , Protected, Private and Access modifiers

class Person():
    def __init__(self, name, age):
        self.name = name ## public veriable 
        self.age = age ## public veraible

def get_name(person):
    return person.name

person = Person('Asim',22)
print(get_name(person))
```

```python
class Person():
    def __init__(self,name,age,gender):
        self._name = name   ## protected variable
        self._age = age     ## protected variable ## acess in the drive class (child) class 
        self._gender = gender ## protected variable it tell only developer not change out of the class

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name,age,gender)
    
def get_name(employee): ## 
    return employee._name

employee = Employee('Asim', 23, "male")
get_name(employee)
```

```python
class Person():
    def __init__(self, name, age):
        self.__name = name  ## private attribute
        self.__age = age    ## private attribute

    ## getter method for name
    def get_name(self):
        return self.__name
    
    ## Getter method for the age
    def get_age(self):
        return self.__age

    ## Setter method for the name
    def set_name(self, name):
        self.__name = name

    ## Setter method for the age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")

## object 
person = Person('Asim',23)

## Access and modify private variable
print(person.get_name())
print(person.get_age())

print(person.set_name("Najim"))
print(person.set_age(20))
print('=================')
print(person.get_name())
print(person.get_age())
print(person.set_age(-5))
```
