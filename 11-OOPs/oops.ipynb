```python
## instance Variable and method

class Dog:
    ## constructor
    def __init__(self,name,age):
        self.name = name
        self.age = age

    ## method
    def getter(self):
        print(self.name, "\n",self.age)

## object
dog1 = Dog('Tommy',4)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2 = Dog('buddy',3)
print(dog2.name)
print(dog2.age)
```

```python
### define a class with instance method
class Dog:

    ## constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    ## instance method
    def bark(self):
        print(f"{self.name} says bhow bhow")

dog1 = Dog('Buddy',3)
dog1.bark()
dog2 = Dog('Tommy', 4)
dog2.bark()
```

```python
## modeling a bank accoutn

## creating a class for the bank account
class BankAccount():
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance

    def deposite(self,amount):
        self.balance += amount
        print(f"{amount} is deposited successfully \nYour balance is {self.balance}")
    def withdraw(self,amout):
        if self.balance < amout:
            print(f"Insufficient balance.\nBalance is  {self.balance}")
        else:
            self.balance -= amout
            print(f"amount is withdraw New balance is {self.balance}")
    def get_balance(self):
        return self.balance
    
## creating  an object
account = BankAccount("Asim",5000)
print(account.balance)
```

```python
## call instance method

account.deposite(5000)
```

```python
account.withdraw(100)
```

```python
account.withdraw(5000)
```
