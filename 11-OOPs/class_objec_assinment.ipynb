### Assignment 1: Basic Class and Object Creation

Create a class named `Car` with attributes `make`, `model`, and `year`. Create an object of the class and print its attributes.


```python
## class
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year

car1 = Car("Toyota",'Fortuner',2019)
print(car1.make,car1.model,car1.year)
```

### Assignment 4: Class with Private Attributes

Create a class named `BankAccount` with private attributes `account_number` and `balance`. Add methods to deposit and withdraw money, and to check the balance. Create an object of the class and perform some operations.


```python
class BankAccount:
    
    def __init__(self,account_noumber,balance=0):
        self.__account_number = account_noumber
        self.__balance = balance

    def deposit(self,amount):
        self.__balance += amount
        print(f"amount {amount} Successfully deposited\nAvailable balance is {self.__balance} ")
    
    def withdrew(self,amount):
        if amount > self.__balance:
            print(f"Insufficeint balance : {self.__balance}")
        else:
            self.__balance -= amount
            print(f"balance withdraw {amount} \nAvailable balance is {self.__balance}")
    
account = BankAccount(23543543,5000)
```

```python
account.deposit(500)
```
