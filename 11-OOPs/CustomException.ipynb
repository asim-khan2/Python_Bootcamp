## Custom Exception (Raise and Throe an excepton)

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter the dob : "))
age = 2024-year

try:
    if age <=30 and age >= 20:
        print("the age is valid so you can apply for the exame")
    else:
        raise dobException
except dobException:
    print("Sorry,your age should be greater then 20 and less than 20")
```
