```python
import logging

## basicCongfig

logging.basicConfig(
    level=logging.DEBUG,
    filename='app.log',
    filemode='w',
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%M-%d %H:%M:%S',
)


## logging message with different severity levels
logging.debug("This is the debug message")
logging.info("This is the info message")
logging.warning("This is the warning")
logging.critical("Thsi is the critical ")
```
