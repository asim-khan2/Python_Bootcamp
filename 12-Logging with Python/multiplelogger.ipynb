#### Logging with Multiple Loggers
You can create multiple loggers for different parts of your application.

```python
import logging 

## create the logger for the module1
logger1 = logging.getLogger('module1')
logger1.setLevel(logging.DEBUG)

## create a logger for module2
logger2 = logging.getLogger('module2')
logger2.setLevel(logging.WARNING)


# Configure logging settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
logger1.debug('this is the debuge for the module1')
logger2.debug('this is the debuge from the module2')
logger2.warning("this is the warning for the module2")
logger2.error("this is the error message form the molule2")
```
