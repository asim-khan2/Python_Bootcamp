```python
# filter function 

# syntax 
# filter(function,list)

# here the function even & odd
def even(num):# even function
    if num % 2 == 0:
        return True
    
def odd(num):
    if num % 2 != 0:
        return True
```

```python
# filter function 

numbers = [1,2,4,6,7,4,8,9,9]

even_number = list(filter(even,numbers)) # return even number from the list
odd_number = list(filter(odd,numbers))

print(f"Even number {even_number}")
print(f"odd number {odd_number}")

print(len(numbers))
```

```python
# filter function with the lamba function 

list(filter(lambda x : x % 2 == 0, numbers))
```

```python
greater_then_five = list(filter(lambda x : x > 5, numbers)) 
print(f"{greater_then_five}")
```
