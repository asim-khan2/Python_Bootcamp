The map() function in Python is used to apply a given function to each item in an iterable (like a list, tuple, etc.) and returns a map object (iterator) that contains the results.

```python
# # map function example
# # syntax 

# map(function, itreble)

numbers = [1,2,3,4,5,6]

list(map(lambda x : x**2, numbers))
```

```python
char = ['2','3','4','6']
print(f"intial list is {char}")

res = list(map(int,char))

print(f"updated list is {res}")
```

```python
names = ['Asim','Khan','Python']

list(map(str.upper,names))
```

```python
a = [1,3,5,5]
b = [5,8,8,9]

res = list(map(lambda a,b : a+b, a,b))
print(res)
```

```python
def double(x):
    return x*x

num = [1,2,4,5,6]

res = map(double,num)
print(list(res))
```

```python
def square(x):
    return x**2

square(3)
```

```python
numb = [1,2,3,6,7,8]

list(map(lambda s : s**2, numb))
```
