```python
# veriable length argument
# positional argument and keywords argument

# def print_number(*Asim): or
def print_number(*args): # positional argument (gives a tuple)
    for number in args:
        print(number)
    
print_number(1,3,5,6,8,64,1.3,'Asim')
```

```python
# keyword argument
# **kwargs to accept any number of keyword arguments and prints each key-value pair
# (gives a dictionary).
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Asim",age=23,city="Amroha")
```

```python
def print_detail(*args,**kwargs):
    for val in args:
        print(f"positional argument : {val}")
    
    for key , val in kwargs.items():
        print(f"{key}:{val}")
```

```python
print_detail(2,4,45,2.2,"Asim",name="Asim",age=23,city="Amroha")
```

```python
# Celsius to Fahrenheit

def Celsius_to_Fahrenheit(c):
    return (c * 9/5) + 32

def Fahrenheit_to_Celsius(F):
    return (F -32) * 5/9

print("Temperature Converter")
print("1 for the Fehrenheit")
print("2 for the Celsius")

choice = input("Enter the choice : ")

if choice == "1":
    c = float(input("Enter temperature in Celsius: "))
    print(f"{c}°C is equal to {Celsius_to_Fahrenheit(c)}°F")

elif choice == "2":
    f = float(input("Enter the temperature in Ferhenheit: "))
    print(f"{f}°F is equal to {Fahrenheit_to_Celsius(f)}°C")

else:
    print('invalid choice Please enter a Valid choice ')
```

```python
# temperature converter

def temp_conversion(temp, unit):
    """this function convert temperature between Celsius to Fehrenheit"""# dok String

    if unit == 'C':
        return (temp * 9/5) + 32
    elif unit == 'F':
        return (temp - 32) * 5/9
    else:
        return None
print(temp_conversion(25,'C'))
print(temp_conversion(77,"F"))

```

```python
# Strong password
def is_strong_password(password):
    if len(password) < 8 :
        return True
    if not any(char.isdigit() for char in password):
        return False

pas = input("Enter the password for the checking strong or not : ")
res = is_strong_password(pas)
if res == True:
    print(f"password is Strong")
elif res == False:
    print(f"password is not Strong")
else:
    print("Please Enter the strong password. Thank You")
```
