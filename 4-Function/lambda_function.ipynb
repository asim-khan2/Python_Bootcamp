A lambda function in Python is a small, anonymous function — meaning it doesn’t have a name like a normal function made with def.

```python
# syntax
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(2,4)
```

```python
addition = lambda a,b : a+b
type(addition)
print(addition(2,3))
```

```python
def even(num):
    if num % 2 == 0:
        return True
```

```python
even(24)
```

```python
even1 = lambda num : num % 2 == 0
```

```python
even1(2)
```

```python
def addition(a,b,c):
    return a+b+c 
```

```python
addition(2,3,4)
```

```python
addition1 = lambda a,b,c : a+b+c
```

```python
addition1(2,3,4)
```

```python
# The map() function applies another function
#  to each item in a list (or any iterable) and
#  returns a new object (a “map object”)
#  which we can convert into a list.

# syntax 
# map(function, iterable)
```

```python
# list 
numbers = [1,2,3,4,5,6]

def number(num):
    return num**2
```

```python
square_numer=lambda x : x**2
```

```python
list(map(number,numbers))
# or
# list(map(lambda x : x**2, numbers))
```
