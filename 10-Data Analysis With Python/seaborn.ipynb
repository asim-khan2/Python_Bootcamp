#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them. 

```python
import seaborn as sns
```

```python
## basic plot 
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
## Scatter plot 
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of Tottal Bill VS tip")
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
## categorical plot
## bar plot

sns.barplot(x='day',y='total_bill',data=tips)
```

```python
## box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## Violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.grid()
```

```python
## histogram 
sns.histplot(x='total_bill',data=tips,kde=True,hue='sex')
```

```python
## kde (kernel density estimetor)
sns.kdeplot(x='total_bill',data=tips,hue='sex',fill=True)
```

```python
## pair plot 
sns.pairplot(tips)
```

```python
tips
```

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
## heat map
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head(3)
```

```python
### plot total sales by product
plt.figure(figsize=(15,10))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,orient='v',)
plt.title('Total sales by product')
plt.xlabel('Categories')
plt.ylabel('Revenue')
plt.show()
```

```python
## plot total sales by region
plt.figure(figsize=(10,8))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,hue='Region')
plt.title('Total sales by Region')
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```
