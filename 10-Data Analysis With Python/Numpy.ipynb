```python
## creating array using numpy
## creae 1D array
import numpy as np
arr1 = np.array([1,3,5,6,5])
print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
## 1D array
arr2 = np.array([1,3,5,6,5])
arr2.reshape((1,5))
# print(arr2.shape)
```

```python
## creating the 2d array
arr2 = np.array([[1,3,5,7,9,5],[3,4,5,3,3,3]])
print(arr2.shape)
print(arr2)
```

```python
## inbuilt method to create the array
np.arange(0,10,2).reshape(5,1)
```

```python
## ones methode 
np.ones((2,5))
```

```python
#### creating the identiy array
np.eye(4)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print("Array:\n",arr)
print("Shape:",arr.shape)
print("Number of dimention : ", arr.ndim)
print("Size of the elements", arr.shape)
print("Data type", arr.dtype)
print("Item Size (in bytes)", arr.itemsize)
```
